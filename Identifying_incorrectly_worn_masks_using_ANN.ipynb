```
# Converting the images to numpy arrays
# Iterating through each folder and converting the images to array to calculate the average image

# Access all jpg files in directory
def avgImg(filepath):
  allimages = os.listdir(filepath)
  print(filepath)
  imlist= [item for item in allimages]
  N=len(imlist)
  # The images are of different sizes. Resizing to (150,150)
  size = (150,150)
  # Create a numpy array of floats to store the average (assume RGB images)
  arr=np.zeros((150,150,3),np.float)

  # Calculating average pixel intensity, converting to float 
  for im in imlist:
      img = Image.open(filepath+im).convert("RGB")
      resized_img = img.resize(size)
      imarr=np.array(resized_img,dtype=np.float)
      arr=arr+imarr/N

  # Round values in array and cast as 8-bit integer
  arr=np.array(np.round(arr),dtype=np.uint8)
  return arr

```

```
# Converting and saving the images to numpy array list
def ImgArr(filepath):
  allimages = os.listdir(filepath)
  print(filepath)
  imlist= [item for item in allimages]
  N=len(imlist)
  # The images are of different sizes. Resizing to (150,150)
  size = (150,150)
  all_images = []
  for im in imlist:
      img = Image.open(filepath+im).convert("L")
      resized_img = img.resize(size)
      imarr=np.array(resized_img)
      all_images.append(imarr)
  return all_images
```

```
from sklearn.decomposition import PCA
from math import ceil

def eigenimages(full_mat,n_comp =0.95):  #95% variability in class
    # fit PCA 
    pca = PCA(n_components = n_comp)
    pca.fit(full_mat)
    print('Number of PC: ', pca.n_components_)    
    return pca
  
```

```
# plot eigenimages in a grid
def plot_pca(pca, title, size = (150,150)):    
    n = 2
    fig = plt.figure(figsize=(8, 6))
    r = int(n**.5)
    c = ceil(n/ r)
    for i in range(n):
        ax = fig.add_subplot(r, c, i + 1, xticks = [], yticks = [])
        ax.imshow(pca.components_[i].reshape(size)
                 )
        ax.set_title(title)
    plt.axis('off')
    plt.show()
```

```
#Converting to array
prp_img_arr = (np.array(prp_img_arr))
imprp_img_arr = (np.array(imprp_img_arr))
```

```
# processing the eigen vectors
prp_eigenfaces = eigenimages(np.array(prp_img_arr).reshape(prp_img_arr.shape[0], 150*150))
imprp_eigenfaces = eigenimages(np.array(imprp_img_arr).reshape(imprp_img_arr.shape[0], 150*150))
```

```
# plotting the various components
plot_pca(prp_eigenfaces,'Proper')
plot_pca(imprp_eigenfaces, 'Improper')

#Blurry images representing different componenets
```

```
#Reconstructing the images to see how all features are combining to form a recognizable image 
components = prp_eigenfaces.transform(prp_img_arr.reshape(prp_img_arr.shape[0], 150*150))
projected = prp_eigenfaces.inverse_transform(components)

components_imp = imprp_eigenfaces.transform(imprp_img_arr.reshape(imprp_img_arr.shape[0], 150*150))
projected_imp = imprp_eigenfaces.inverse_transform(components_imp)
```

```
# plot eigenimages in a grid
def plot_pca_recon(projected, title, size = (150,150)):    
    n = 4
    fig = plt.figure(figsize=(16, 10))
    r = int(n**.5)
    c = ceil(n/ r)
    for i in range(4):
        ax = fig.add_subplot(r, c, i + 1, xticks = [], yticks = [])
        ax.imshow(projected[i].reshape(size)
                 )
        ax.set_title(title)
    plt.axis('off')
    plt.show()
```

```
#Visualizing the combined features from PCA
plot_pca_recon(projected,'Proper')
plot_pca_recon(projected_imp, 'Improper')
```

## CNN - Model 4 - VGG16 - Fine Tuning ##





We have used VGG16 in Model3, but here we freeze the convolutional base before we compile and train our model. "Freezing" a layer or set of layers means preventing their weights from getting updated during training. 
We will `fine-tune` only certain convolutional layers to make it more relevant to our specific data. 

### Transform the data ###

```
conv_base_r = VGG16(weights='imagenet', # weights from imagenet - very important! otherwise it's just the network structure
                  include_top=False, # this means that our dense layer will be trained specific to our small data problem that we are solving
                  input_shape=(150, 150, 3)) # this is the shape of the data from our data generator 
```

```
# fine tuning the base VGG16 
# Replacing the last 3 layers 

conv_base_r.trainable = True

set_trainable = False
for layer in conv_base_r.layers:
    if layer.name == 'block5_conv1':
        set_trainable = True
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False
```

```
conv_base_r.summary()
```

```
train_datagen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')

#validation data should not be augmented!
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=10,
        class_mode='binary',shuffle=False)

```

### Fit Model ###

```
model4 = models.Sequential()
model4.add(conv_base_r)
model4.add(layers.Flatten())
model4.add(layers.Dense(256, activation='relu'))
model4.add(layers.Dense(1, activation='sigmoid'))
```

```
model4.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-5), # low learning rate is the magnitude of the modifications as result of fine tuning
              metrics=['acc'])

history4 = model4.fit_generator(
      train_generator,
      steps_per_epoch=35,
      epochs=20,
      validation_data=validation_generator,
      validation_steps=15,
      verbose=2)

pd.DataFrame.from_dict(history4.history).to_csv('/content/drive/My Drive/OPIM 5509 - Deep Learning/Module 3/Project/Models/history.csv',index=False)
```
